````python
#Create a list of 5 models with name, accuracy, speed, and params.
#Write a function called best_small_model that:

#Loops through all models
#Filters only models with accuracy above 0.92 AND params below 100
#Among those filtered models, returns the one with the highest accuracy
#Prints: "Best small model: MiniLM with 95% accuracy"
````

````python
models = [
    {"name": "XGBoost", "accuracy": 0.94, "speed": "fast", "params": 50},
    {"name": "GPT-4", "accuracy": 0.98, "speed": "slow", "params": 1700000},
    {"name": "MiniLM", "accuracy": 0.95, "speed": "fast", "params": 22},
    {"name": "ResNet", "accuracy": 0.91, "speed": "slow", "params": 25},
    {"name": "TinyBERT", "accuracy": 0.94, "speed": "fast", "params": 67}
]


def best_small_model (models):


  newModelAcc = []
  for model in models:
    if model["accuracy"] > 0.92 and model["params"] < 100:
      newModelAcc.append(model)

  best = newModelAcc[0]

  for model in newModelAcc:
    if model["accuracy"] > best["accuracy"]:
      best = model

  print(f"Best small model: {best["name"]} with {best["accuracy"]: .0%} accuracy")

  return best


best_small_model(models)
````

````python
#**Exercise 2 — `**kwargs`:**
#Write a function called `build_report` that accepts any keyword arguments and prints a formatted report like:
##```
#===== Report =====
#name: XGBoost
#accuracy: 94%
#speed: fast
#params: 50M
#==================
#Accuracy should be formatted as percentage, params should have M after it, everything else prints normally.
````

````python
def build_report (**kwargs):

  print(f"===== Report =====")

  for key, value in kwargs.items():
    if key == "accuracy":
      print(f"{key}: {value: .0%}")
    elif key == "params":
      print(f"{key}: {value}M")
    else:
      print(f"{key}: {value}")

  print(f"==================")
````

````python
#Step 1 — Build the parent class AIModel:

#Takes name, accuracy, speed in __init__
#Method describe — prints: "XGBoost | 94% | fast"
#Method is_production_ready — returns True if accuracy above 0.92 AND speed is "fast"

#Step 2 — Build 2 child classes:
#FinanceModel(AIModel):

#Adds bank_name in __init__
#Method risk_check — prints: "Chase — FraudDetector risk level: LOW" if accuracy above 0.95, otherwise "HIGH"

#HealthModel(AIModel):

#Adds hospital_name in __init__
#Method safety_check — prints: "Mayo Clinic — DiagnosisAI safety: APPROVED" if accuracy above 0.95 AND speed is fast, otherwise "REJECTED"

#Step 3 — Create instances and test:
#python# Finance models
#f1 = FinanceModel("FraudDetector", 0.96, "fast", "Chase")
#f2 = FinanceModel("CreditScorer", 0.89, "slow", "Wells Fargo")

# Health models
#h1 = HealthModel("DiagnosisAI", 0.97, "fast", "Mayo Clinic")
#h2 = HealthModel("XRayAnalyzer", 0.94, "slow", "Johns Hopkins")
````

````python
class AIModel:
  def __init__(self, name, accuracy, speed):
    self.name = name
    self.accuracy = accuracy
    self.speed = speed

  def describe (self):
    print(f"{self.name} | {self.accuracy :.0%} | {self.speed}")

  def is_production_ready (self):
    return(self.accuracy > 0.92 and self.speed == "fast")

#child class 1

class FinanceModel(AIModel):
  def __init__(self, name, accuracy, speed, bank_name):
    super().__init__(name, accuracy, speed)
    self.bank_name = bank_name

  def risk_check(self):
    #prints: "Chase — FraudDetector risk level: LOW" if accuracy above 0.95, otherwise "HIGH"

    if self.accuracy >0.95:
      print(f"{self.bank_name} - {self.name} risk level: LOW")
    else:
      print(f"{self.bank_name} - {self.name} risk level: HIGh")

#HealthModel(AIModel):

#Adds hospital_name in __init__
#Method safety_check — prints: "Mayo Clinic — DiagnosisAI safety: APPROVED" if accuracy above 0.95 AND speed is fast, otherwise "REJECTED"

class HealthModel(AIModel):

  def __init__(self, name, accuracy, speed, hospital_name):
    super().__init__(name, accuracy, speed)
    self.hospital_name = hospital_name

  def safety_check (self):

    if self.accuracy >0.95 and self.speed == "fast":
      print(f"{self.hospital_name} - {self.name} safety: APPROVED")
    else:
      print(f"{self.hospital_name} - {self.name} safety: REJECTED")

#Financial Models
f1 = FinanceModel("FraudDetector", 0.96, "fast", "Chase")
f2 = FinanceModel("CreditScorer", 0.89, "slow", "Wells Fargo")

# Health models
h1 = HealthModel("DiagnosisAI", 0.97, "fast", "Mayo Clinic")
h2 = HealthModel("XRayAnalyzer", 0.94, "slow", "Johns Hopkins")

f1.describe()
print(f1.is_production_ready())
f1.risk_check()

f2.describe()
print(f2.is_production_ready())
f2.risk_check()

h1.describe()
print(h1.is_production_ready())
h1.safety_check()

h2.describe()
print(h2.is_production_ready())
h2.safety_check()
````
